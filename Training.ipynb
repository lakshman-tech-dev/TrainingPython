```python
s1 = input("Enter first string: ")
s2 = input("Enter second string: ")

if len(s1) != len(s2):
    print("Not Anagram")
else:
    flag = True
    for ch in s1:
        if s1.count(ch) != s2.count(ch):
            flag = False
            break
    if flag:
        print("Anagram")
    else:
        print("Not Anagram")
```

```python
s = input()
vowels = "aeiouAEIOU"
v = []
for ch in s:
    if ch in vowels:
        v.append(ch)
v.reverse()
result = ""
j = 0
for ch in s:
    if ch in vowels:
        result += v[j]
        j += 1
    else:
        result += ch
print(result)
```

```python
s = input()
longest = 0
for i in range(len(s)):
    temp = ""
    for j in range(i, len(s)):
        if s[j] not in temp:
            temp += s[j]
            longest = max(longest, len(temp))
        else:
            break
print(longest)
```

```python
s = input()
longest = ""
for i in range(len(s)):
    for j in range(i, len(s)):
        sub = s[i:j+1]
        if sub == sub[::-1]:
            if len(sub) > len(longest):
                longest = sub
print(longest)
```

```python
n = int(input())

rev = 0
temp = n

while temp != 0:
    digit = temp % 10
    rev = rev * 10 + digit
    temp //= 10

count1 = 0
for i in range(1, n + 1):
    if n % i == 0:
        count1 += 1
count2 = 0
for i in range(1, rev + 1):
    if rev % i == 0:
        count2 += 1

if count1 == 2 and count2 == 2:
    print("Prime")
else:
    print("Not prime")
```
